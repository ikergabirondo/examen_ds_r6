```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv('netflix_titles.csv').head(100)
```

```python
df.head()
```

## Recomendador de películas (2,5 puntos)

```python
dfMovies = df[df['type']=='Movie']
```

```python
dfMovies
```

```python
dfMovies = dfMovies.reset_index()
```

```python
dfMovies2 = dfMovies.reset_index()
dfMovies2
```

```python
dfMovies2[['director','cast','listed_in','description']] = dfMovies2[['director','cast','listed_in','description']].fillna('')
```

```python
dfMovies2.isnull()
```

```python
def combined_features(row):
    return str(row['director'])+" "+str(row['cast'])+" "+str(row['listed_in'])+" "+str(row['description'])
dfMovies["combined_features"] = df.apply(combined_features, axis =1)
```

```python
dfMovies['combined_features']
```

```python
#A continuación creamos la matriz de similitud

vectorizer = CountVectorizer().fit_transform(dfMovies["combined_features"].values)
vectors = vectorizer.toarray()
csim = cosine_similarity(vectors)
```

```python
csim
```
